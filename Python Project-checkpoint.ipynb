```python
# Below is calculate for Cost Price, Selling Price and Profit
sales['Cost_Price'] = sales['Order_Quantity'] * sales['Unit_Cost']
sales['Selling_Price'] = sales['Order_Quantity'] * sales['Unit_Price']
sales['Profit'] = sales['Selling_Price'] - sales['Cost_Price']
sales
```

Convert the date column to datetime format

```python
sales['Date'] = pd.to_datetime(sales['Date'])
```

```python
# Extract year, month, and day
sales['Year'] = sales['Date'].dt.year
sales['Month'] = sales['Date'].dt.month_name()
sales['Day'] = sales['Date'].dt.day_name()
sales
```

#1 Use the IF Function to group the age column into 4 categories (< 25 as Youth, 
25 – 34 as Young Adult, 35 – 64 as Adults and 64+ as Senior)

```python
# Using conditional formating for grouping age group and name the colunm as Age_Group
sales['Age_Group'] = [
    'Youth' if age < 25 else
    'Young Adult' if age <= 34 else
    'Adult' if age <= 64 else
    'Senior'
    for age in sales['Customer_Age']
]
sales
```

#2 Which country raked in the highest profit? (Use an apt chart to show this)

```python
# Using group by to group the country column and sort the value in decending order.
sales.groupby ('Country')['Profit'].sum().sort_values(ascending = False)
```

```python
# Importing matplotlib to visualize the above query.
import matplotlib.pyplot as plt
```

```python
# Visual for chart
plt.figure(figsize = (12, 3))
sales.groupby ('Country')['Profit'].sum().sort_values(ascending = False).plot(kind = 'bar', color = 'Green')
plt.title('Profit by Country', fontsize = 20)
plt.ylabel('Profit')
plt.xticks(rotation = 360);
```

#3 Which age group patronized the company most? (Use quantity ordered as metric)

```python
# Grouping by age group and total sum of quantity order in a desending order
sales.groupby('Age_Group')['Order_Quantity'].sum().sort_values(ascending = False)
```

```python
# Chart visual
plt.figure(figsize = (12, 5))
sales.groupby ('Age_Group')['Order_Quantity'].sum().sort_values(ascending = False).plot(kind = 'bar')
plt.xlabel('Age Group')
plt.ylabel('Order Quantity')
plt.title('Order by Age Group', fontsize = 16)
plt.xticks(rotation = 0);
```

#4. Which product was sold most in Canada and United States? (Use an apt chart 
to show this)

```python
# Filter Canada and United Sates from country column and asign canada_us_sales to them
canada_us_sales = sales[(sales['Country'] =='United States') | (sales['Country'] == 'Canada')]
canada_us_sales
```

```python
# Group canada_us_sales by product and return the sum of order quantity and sort values in descending order and filter to top 5
canada_us_sales.groupby('Product')['Order_Quantity'].sum().sort_values(ascending = False).iloc[ :5]
```

```python
# Chart Visualization
plt.figure(figsize = (12, 3))
canada_us_sales.groupby('Product')['Order_Quantity'].sum().sort_values(ascending = False).iloc[ :5].plot(kind = 'bar', color = 'purple')
plt.title('5 Most sold Product both in USA and Canada', fontsize = 10)
plt.xlabel('Product')
plt.ylabel('Order_Quantity')
plt.xticks(rotation = 45);
```

#5 What is the least sold product? (Use an apt chart to show this)

```python
# Group by product and sum by quantity order and sort value in ascending order
sales.groupby('Product')['Order_Quantity'].sum().sort_values(ascending = True)
```

```python
# Chart visualization
plt.figure(figsize = (8, 3))
sales.groupby('Product')['Order_Quantity'].sum().sort_values(ascending = True).iloc[:5].plot(kind = 'bar', color = 'purple')
plt.title('Bottom 5 Products', fontsize = 10)
plt.xlabel('Product')
plt.ylabel('Order_Quantity')
plt.xticks(rotation = 45);
```

#7 Show trend of profit by Year, Month and Day. Which actionable insight do you 
notice? (Use an apt chart to show this)


```python
# Date time have been imported during date clearing above, so we group by year, month and day
sales.groupby('Year')['Profit'].sum()
```

To visualize trend of profit by Years,Months and Days, we use line chat and we import seaborn to make our line chat.

```python
import seaborn as sns
```

```python
# Assign year to the variable on X-axis as index
Year = sales.groupby('Year')['Profit'].sum().index
Year
```

```python
# Assign profit to the varible on y-axis as values
Profit = sales.groupby('Year')['Profit'].sum().values
Profit
```

```python
# Chart visualization for sales trend over the year
plt.figure(figsize = (10, 3)) 
sns.lineplot(x = Year, y = Profit);
plt.xticks(rotation = 0);
```

```python
# Assign month to the variable on X-axis as index
Month = sales.groupby('Month')['Profit'].sum().index
Month
```

```python
# Assign month to the variable on y-axis as value
Profit = sales.groupby('Month')['Profit'].sum().values
Profit
```

```python
# Chart visualization for sales trend over the year
plt.figure(figsize = (10, 5)) 
sns.lineplot(x = Month, y = Profit)
plt.xticks(rotation = 100);
```

```python
# Assign day to the variable on x-axis as index
Day = sales.groupby('Day')['Profit'].sum().index
Day
```

```python
# Assign profit to the variable on y-axis as value
Profit = sales.groupby('Day')['Profit'].sum().values
Profit
```

```python
# Chart visualization for sales trend over the year
plt.figure(figsize = (10, 5)) 
sns.lineplot(x = Day, y = Profit)
plt.xticks(rotation = 100);
```
